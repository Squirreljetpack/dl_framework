```python
# allow interaction with the plots
%matplotlib widget

%load_ext autoreload
%autoreload 2

# add the main directory reference and import 'imports.py'
import sys

if ".." not in sys.path:
    sys.path.insert(0, "..")
from imports import *

logging.basicConfig(level=logging.ERROR)
__builtins__.verbosity = 5

torch.set_printoptions(precision=2)


class params:
    batch_size = 256
    num_workers = 16
    max_epochs = 100
    seq_len = 48
    lr = 0.01
```

```python
from data.AQ import *

dt = AQ(AQConfig.create(params))
dt.preview(samples=1)
sample = dt.sample_batch()[0]
```

```python
from models.RNN import *
from lib.train import *

rnn_config = RNNConfig(features=sample.shape[-1], hidden_size=64, layers=2)
trainer_config = TrainerConfig.create(
    params,
)
```

```python
plt.close("all")

model = RNN(rnn_config)

trainer = Trainer(trainer_config)
trainer.fit(model, dt.loaders())
```

```python
plt.close("all")

trainer = Trainer(trainer_config)

loss_board = make_loss_board()

for i in dt.folds():  # dt was configured with 6-fold splits
    model = RNN(rnn_config)
    print(dt.data_inds)
    trainer.init(model, dt.loaders(), loss_board=loss_board)
    trainer.train_loss_mf = None
    trainer.val_loss_mf.rename("loss", f"loss {i}")
    trainer.fit()
```

```python
plt.close("all")

trainer = Trainer(trainer_config)

loss_board = make_loss_board()

for flavor in ["rnn", "lstm", "gru"]:  # dt was configured with 6-fold splits
    model = RNN(rnn_config.update(flavor=flavor))
    trainer.init(model, dt.loaders(), loss_board=loss_board)
    trainer.train_loss_mf = None
    trainer.val_loss_mf.rename("loss", f"loss ({flavor})")
    trainer.fit()
```

```python
from lib.wandb import *

plt.close("all")
trainer_config.max_epochs = 1

obj = make_objective(
    RNN,
    rnn_config,
    dt.loaders,
    trainer_config,
    from_te(torcheval.metrics.MeanSquaredError, "mse"),
)

sweep_configuration = {
    "method": "bayes",  # Random search method (alternatives: 'grid', 'bayesian')
    "name": "rnn_hyperparam_tuning",
    "metric": {
        "name": "mse",
        "goal": "minimize",
    },  # used for bayesian search, not random
    "parameters": {
        "lr": {"min": 1e-5, "max": 1e-1},
        "batch_size": {"values": [16, 32, 64, 128]},
        "hidden_size": {"values": [64, 128, 256, 512]},
        "dropout": {"min": 0.1, "max": 0.5},
    },
}

tune(sweep_configuration, obj, 5)
```
